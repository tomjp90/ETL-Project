# Loop through suburbs and obtain counts on search terms

# WARNING! This code calls google places API and takes over 10mins to run.

```python
    start = time.time()   
    
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

    search_terms = ["liquor_store", "police", "park"]

    for term in search_terms:
        print(f"{term} -------------------------------------------")
        
        for index, row in melb_suburbs_df.iterrows():
            
            longitude = melb_suburbs_df["lon"][index]
            latitude = melb_suburbs_df["lat"][index]
            suburb = melb_suburbs_df["suburb"][index]
            location = f"{latitude},{longitude}"

            # parameters for search
            params = {"location": location, 
                      "radius": 1000, 
                      "types": term,
                      "key": gkey
                     
                     }

            # Search and retrieve json from query request
            response = requests.get(base_url, params)
            response_json = response.json()
            
            print(f"searching {suburb} for {term}")

            try:
                # put the result into column
                melb_suburbs_df.loc[index, f"No {term}"] = len(response_json["results"])

            except:
                # if no result found print nothing found
                melb_suburbs_df.loc[index, f"No. {term}"] = 0         
            
end = time.time()
time_taken = end - start
print("------------------------------------------------------------")
print(f"API requests have taken {time_taken} seconds" )
melb_suburbs_df.to_csv('01_Outputs/melb_suburbs_types.csv', index=False)
```
