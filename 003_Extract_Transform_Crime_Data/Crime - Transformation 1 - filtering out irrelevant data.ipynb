```python
import pandas as pd
```

```python
#Define path to downloaded Excel for crime data
crimeratepath = "../000_Resources/Crime rate/Data_Tables_LGA_Criminal_Incidents_Year_Ending_September_2020.xlsx"
```

```python
#Create dataframe for crime data - Can take a couple of minutes to read
crimerateDF = pd.read_excel(crimeratepath, sheet_name = "Table 03")
crimerateDF
```

```python
#As suburbs had duplicate postcodes, merging by postcodes caused issues, Determined best to merge via suburbs
#Needed to format crime dataframe in preparation
crimerateDF2 = crimerateDF.rename(columns={"Suburb/Town Name":"suburb"})
crimerateDF2['suburb'] = crimerateDF2['suburb'].str.lower()
crimerateDF2
```

```python
#Define path to suburb postcode file
suburb_postcodes_path = "../001_Extract_Transform_Suburbs/01_Outputs/melbourne_suburbs_postcode.csv"
```

```python
#Read in Excel and create DF for suburb postcode file
suburb_postcodes_DF = pd.read_csv(suburb_postcodes_path)
suburb_postcodes_DF
```

```python
#Merging suburbs and crime data so crime data only includes the suburbs in the suburbs file
crime_suburb_merge2 = pd.merge(suburb_postcodes_DF, crimerateDF2, on="suburb",  how ="left")
crime_suburb_merge2
```

```python
#Check unique values in suburb
ChecksyburbuniqueABC = crime_suburb_merge2["suburb"].unique()
ChecksyburbuniqueABC.sort()
len(ChecksyburbuniqueABC)
```

```python
#Check unique values in state
CheckstateuniqueABC = crime_suburb_merge2["state"].unique()
CheckstateuniqueABC.sort()
print(CheckstateuniqueABC)
len(CheckstateuniqueABC)
```

```python
#Check unique values in LGA
CheckLGAuniqueABC = crime_suburb_merge2["Local Government Area"].unique()
CheckLGAuniqueABC.sort()
print(CheckLGAuniqueABC)
len(CheckLGAuniqueABC)
```

```python
#Drop unneccesary columns
crime_suburb2 = crime_suburb_merge2[['Year', 'Year ending', 'Local Government Area', 'Postcode',
       'suburb', 'Offence Division', 'Offence Subdivision',
       'Offence Subgroup', 'Incidents Recorded']]
crime_suburb2
```

```python
#To aggregate lowest categetory of offences into nxxt category up, grouping is required
crimegrouping = crime_suburb2.groupby(['Year', 'Local Government Area', "Postcode", 'suburb', 
                                         'Offence Subdivision'])
```

```python
#Add all related offence subdivisons by year
CrimeSum = crimegrouping['Incidents Recorded'].sum()
```

```python
#Convert to dataframe - mult-index
CrimeSumDFindex = pd.DataFrame(CrimeSum)
CrimeSumDFindex
```

```python
#Convert multi-index dataframe to single-index
CrimeSumDFreset = CrimeSumDFindex.reset_index(level=['Year', 'Local Government Area', 'Postcode', 
                                         'suburb', 'Offence Subdivision'])
CrimeSumDFreset
```

```python
#Next step is to find average for each offence from 2011-2020
#start by grouping
yeargrouping = CrimeSumDFreset.groupby(['Local Government Area', "Postcode", 'suburb', 
                                         'Offence Subdivision'])
```

```python
#Find the mean of the incidents recorded
yearAverage = yeargrouping['Incidents Recorded'].mean()
```

```python
#Convert to dataframe - mult-index
yearAverageDFindex = pd.DataFrame(yearAverage)
yearAverageDFindex
```

```python
#Convert multi-index dataframe to single-index
yearAverageDFreset = yearAverageDFindex.reset_index(level=['Local Government Area', 'Postcode', 
                                         'suburb', 'Offence Subdivision'])
yearAverageDFreset
```

```python
# split the code from the offence name in the Offence Subdivision column
crimeID_DF = yearAverageDFreset["Offence Subdivision"].str.split(" ", n = 1, expand = True)
crimeID_DF
```

```python
# Remove original column name 'Offence Subdivision' 

remove_column = yearAverageDFreset.drop(columns=['Offence Subdivision'])

remove_column
```

```python
# making offence ID column from crimeID_DF
remove_column["crime_id"] = crimeID_DF[0] 

remove_column
```

```python
# making new offence name column from crimeID_DF
remove_column["crime_name"] = crimeID_DF[1] 

remove_column
```

```python
#Re-ordering columns into logical order
crime_DF_final = remove_column[["Local Government Area", "Postcode", "suburb", "crime_id", "crime_name", "Incidents Recorded"]]
crime_DF_final
```

```python
#rename columns to clean up and correctly labelled
crime_DF_final = crime_DF_final.rename(columns={"Postcode":"postcode", "Incidents Recorded":"average_incdent"})
crime_DF_final
```

```python
#Export to CSV
crime_DF_final.to_csv('01_Outputs/crime_DF.csv', index=False)
```
