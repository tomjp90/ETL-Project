```python
import pandas as pd
import numpy as np
import pandas as pd
```

```python
# define path to Heesu's final suburb list
suburbs_path = "01_Outputs/melbourne_suburbs.csv"
```

```python
# define path to downloaded postcode lat long 's list
postcode_path = "../000_Resources/Aust Postcode Location/Australian_Post_Codes_Lat_Lon.csv"
```

```python
#read in csv for final suburbs list
suburbnames = pd.read_csv(suburbs_path)
```

```python
#read in csv for postcode lat long list
postcode = pd.read_csv(postcode_path)
```

```python
#Dataframes for final suburbs list
suburbnamesDF = pd.DataFrame(suburbnames)
#Convert column name all merge
suburbnamesDF['suburb'] = suburbnamesDF['suburb'].str.lower()
suburbnamesDF
```

```python
#Dataframes for postcode long lat list
postcodeDF = pd.DataFrame(postcode)
#Dataframes for final suburbs list
postcodeDF['suburb'] = postcodeDF['suburb'].str.lower()
postcodeDF
```

```python
#Merge on suburb
suburbpostcodemerge = pd.merge(suburbnamesDF, postcodeDF, on= "suburb",  how ="left")
suburbpostcodemerge
```

```python
# Get names of indexes for which column state doesn not equal VIC
indexNames = suburbpostcodemerge[ suburbpostcodemerge['state'] != "VIC" ].index
```

```python
# Delete these row indexes from dataFrame
suburbpostcodemerge.drop(indexNames , inplace=True)
```

```python
#Check DF
suburbpostcodemerge
```

```python
#Check datatype for suburns 
suburbpostcodemerge['postcode'].dtypes
```

```python
#convert suburbs to integer
suburbpostcodemerge.postcode = suburbpostcodemerge.postcode.astype(int)
```

```python
#strip leading and trailing space
suburbpostcodemerge['type'] = suburbpostcodemerge['type'].str.strip()
```

```python
#Check uniqie values in suburb
suburbpostcodeunique = suburbpostcodemerge["suburb"].unique()
suburbpostcodeunique.sort()
print(suburbpostcodeunique)
len(suburbpostcodeunique)
```

```python
#Determined there are duplicates as suburbs and postcodes included for PO Boxes and Delivery Centres

indexNames2 = suburbpostcodemerge[suburbpostcodemerge['type'] != "Delivery Area" ].index
```

```python
# Delete these row indexes from dataFrame
suburbpostcodemerge.drop(indexNames2 , inplace=True)
```
