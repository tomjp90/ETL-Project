```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

# Greater Melbourne suburbs 

  ### Using requests library to get a response from website

```python
url = "https://www.domain.com.au/news/melbournes-321-suburbs-ranked-for-liveability-20151106-gkq447/#:~:text=Melbourne's%20321%20suburbs%20ranked%20for%20liveability"
response = requests.get(url)
response
```

 ### creating beautifulsoup object 

```python
soup = BeautifulSoup(response.text, 'lxml')
soup
```

 ### searched through website code to locate suburb names

```python
results = soup.find("div", class_ = "domain-article-content-wrap").find_all('h3')
```

```python
results
```

 ### creating a list of suburbs and cleaning data by removing html tags 

```python
suburb_list = []

suburb_list = [result.text for result in results]
suburb_list
```

### creating dataframe with suburbs list 

```python
suburbs_df = pd.DataFrame({'suburb':suburb_list})
```

```python
suburbs_df
```

```python
# remove erroneous suburb
suburbs_df = suburbs_df.loc[suburbs_df["suburb"]!="St Helena",:]
suburbs_df
```

```python
suburbs_df.to_csv('01_Outputs/melbourne_suburbs.csv', index=False)
```
